# cisco sales data analysis

```python
## import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.graph_objects as go
from sklearn.model_selection import train_test_split
import warnings
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.arima_model import ARIMA
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
```

```python
##read data 
data=pd.read_csv("CISCO(Cisco Systems).csv")

#first five row of data 
data.head()
```

```python
data['Date']=pd.to_datetime(data['Date'])
```

```python
data.info()
```

```python
import matplotlib.dates as mdates

data['Date'] = pd.to_datetime(data['Date'])
plt.figure(figsize=(19,7))
sns.lineplot(data=data, x='Date', y='Close', linewidth=2.5)
plt.grid(True)
plt.title('CISCO Close Price')
plt.gca().xaxis.set_major_locator(mdates.MonthLocator(interval=3))  # every 3 months
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
def check_stationary(timeseries):
    roll_mean = timeseries.rolling(12).mean()
    roll_std = timeseries.rolling(12).std()
    
    plt.figure(figsize=(17,7))
    plt.plot(timeseries, color='blue', label='Original')
    plt.plot(roll_mean, color='red', label='Rolling Mean')
    plt.plot(roll_std, color='black', label='Rolling Standard Deviation')
    plt.title('Rolling Mean and Standard Deviation')
    plt.legend()
    plt.show()
```

```python
check_stationary(data.Close)
```

```python
##checking adfuller test
adft=adfuller(data.Close,autolag='AIC')
output= pd.Series(adft[0:4],index=['Test Statistics','p-value','No. of lags used','Number of observations used'])
for key,values in adft[4].items():
        output['critical value (%s)'%key] =  values
        print(output)
```

```python
## calculate moving average 
data['ma']=data['Close'].rolling(12,min_periods=1).mean()
```

```python
plt.figure(figsize=(15,8))
plt.plot(data.Date, data.ma)
plt.title('Moving Average over Time')
plt.xlabel('Date')
plt.ylabel('Moving Average')
plt.grid(True)
plt.show()
```

```python
##split the data into train and test
x_train,x_test,y_train,y_test=train_test_split(data['Close'],data['ma'],test_size=0.2)
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
x_train=np.array(x_train)
x_test=np.array(x_test)
y_train=np.array(y_train)
y_test=np.array(y_test)
```

```python
x_train=x_train.reshape(-1,1)
x_test=x_test.reshape(-1,1)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor()
rf.fit(x_train,y_train)
```

```python
rf_pred=rf.predict(x_test)
```

```python
print("Mean Absolute Error:", mean_absolute_error(y_test, rf_pred))
print("Coefficient of Determination:", r2_score(y_test, rf_pred))
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
lr_pred=lr.predict(x_test)

print("Mean Absolute Error:", mean_absolute_error(y_test, lr_pred))
print("Coefficient of Determination:", r2_score(y_test, lr_pred))
```
